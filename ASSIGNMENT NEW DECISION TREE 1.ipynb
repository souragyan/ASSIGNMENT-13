```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import  DecisionTreeClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
data=pd.read_csv('Company.csv',index_col=0)
```

```python
data.columns
```

```python
data.head()
```

```python
data.shape
```

```python
data[data.isnull().any(axis=1)].head()
```

```python
df=data.dropna()
```

```python
dummies=pd.get_dummies(df.ShelveLoc)
dummies.head(2)
```

```python
df=pd.concat([df,dummies],axis='columns')
df.head(2)
```

```python
dummies=pd.get_dummies(df.Urban)
dummies.head(2)
```

```python
df=pd.concat([df,dummies],axis='columns')
df.head(2)
```

```python
dummies=pd.get_dummies(df.US)
dummies.head(2)
```

```python
df=pd.concat([df,dummies],axis='columns')
df.head(2)
```

```python
dummies=pd.get_dummies(df.Sale)
dummies.head(2)
```

```python
df=pd.concat([df,dummies],axis='columns')
df.head(2)
```

```python
df.drop('ShelveLoc',axis='columns',inplace=True)
df.drop('Urban',axis='columns',inplace=True)
df.drop('US',axis='columns',inplace=True)
df.drop('Sale',axis='columns',inplace=True)
df.drop('BAD',axis='columns',inplace=True)
```

```python
df
```

```python
x=df.iloc[:,0:14]
y=df.iloc[:,14]
```

```python
# Splitting data into training and testing data set
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.33,random_state=324)
```

```python
model=DecisionTreeClassifier(max_leaf_nodes=10,random_state=0)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
confusion_matrix(y_test,y_pred)
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.25,random_state=300)
model=DecisionTreeClassifier(max_leaf_nodes=10,random_state=0)
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)*100
```

```python
confusion_matrix(y_test,y_pred)
```
